```python
from sklearn.svm import SVC

X = np.copy(X_train)
y = np.copy(ravel(y_train))

print(shape(X))
print(shape(y))

# SVM Classifier model
svm_clf = SVC(kernel="linear", C=float(150))
svm_clf.fit(X, y)
```

```python
xlim1 = min(X_pca[:,0])
xlim2 = max(X_pca[:,0])

fig = plt.figure(1, figsize=(10, 6))

def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    plt.subplot(121)
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    # At the decision boundary, w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    svs = svm_clf.support_vectors_
    support_indices = svm_clf.support_
    for i in range(len(support_indices)):
        plt.scatter(X_pca[support_indices[i], 0],
            X_pca[support_indices[i], 1], s=100, facecolors='#CFAAAA')
    plt.scatter(svs[:, 0], svs[:, 1], s=200, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)
    print(svs)
    print(svm_clf.support_)

plot_svc_decision_boundary(svm_clf, xlim1, xlim2)
for i in range(split_index):
    plt.subplot(121)
    plt.scatter(X_pca[i, 0], X_pca[i, 1], c=label_colours[i],
           edgecolor='k')
plt.title("PCA with decision boundary", fontsize=16)
plt.show()

xlim1 = min(X_pca_scaled[:,0])
xlim2 = max(X_pca_scaled[:,0])

def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    plt.subplot(122)
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]

    # At the decision boundary, w0*x0 + w1*x1 + b = 0
    # => x1 = -w0/w1 * x0 - b/w1
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0]/w[1] * x0 - b/w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin

    svs = svm_clf.support_vectors_
    support_indices = svm_clf.support_
    for i in range(len(support_indices)):
        plt.scatter(X_pca_scaled[support_indices[i], 0],
            X_pca_scaled[support_indices[i], 1], s=100, facecolors='#CFAAAA')
    plt.scatter(svs[:, 0], svs[:, 1], s=200, facecolors='#FFAAAA')
    plt.plot(x0, decision_boundary, "k-", linewidth=2)
    plt.plot(x0, gutter_up, "k--", linewidth=2)
    plt.plot(x0, gutter_down, "k--", linewidth=2)
    print(svs)
    print(svm_clf.support_)

plot_svc_decision_boundary(svm_clf, xlim1, xlim2)
for i in range(split_index):
    plt.subplot(122)
    plt.scatter(X_pca_scaled[i, 0], X_pca_scaled[i, 1], c=label_colours[i],
           edgecolor='k')
plt.title("Scaled", fontsize=16)
plt.show()
```
